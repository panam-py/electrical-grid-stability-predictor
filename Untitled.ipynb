Importing the tools

```python
# EDA Tools
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Making the plots visible in the notebook
%matplotlib inline

# Modelling Tools
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score, classification_report, plot_roc_curve, plot_confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier
from xgboost import XGBClassifier
from joblib import dump, load
```

**Getting the data ready for modelling**

```python
# Converting the stability column to 0 for stable and 1 for unstable
stability_list = []
for i in df["stabf"]:
    if i == "stable":
        stability_list.append(0)
    else:
        stability_list.append(1)
```

```python
len(stability_list), len(df["stabf"])
```

```python
# Adding the transformed stability column to the dataframe
df.drop(["stabf"], axis=1, inplace=True)
df["stability"] = stability_list
```

```python
df.info()
```

## Modelling

```python
# Splitting the data into data and labels
X = df.drop(["stability"], axis=1)
y = df["stability"]
```

```python
# Splitting X and y into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Creating a function to train and test the models
def fit_score(estimators, train_data, train_labels, test_data, test_labels):
    """
    Trains and tests models, returns a dictionary containing the accuracy scores of the trained models and the predictions of the models
    estimators : A dictionary containing the models to be trained
    train_data : The data to be used to train the models (without labels)
    train_labels : The labels for thr training data
    test_data : The data to be used to test the trained models (without labels)
    test_labels : The labels for the test data
    """
    
    # Setting up a random seed to make results reproducible
    np.random.seed(7)
    
    # Creating a dictionary to store the accuracy scores of the trained models
    scores = {}
    predictions = {}
    
    # Looping through the estimators dictionary to train and test each model
    for name, model in estimators.items():
        model.fit(train_data, train_labels)
        prediction = model.predict(test_data)
        score = model.score(test_data, test_labels)
        
        scores[name] = f"{score * 100:.2f}"
        predictions[name] = prediction
        
    return scores, predictions
```

```python
# Instantiating the models
rf = RandomForestClassifier()
cat = CatBoostClassifier()
xgb = XGBClassifier()

# Creating a dictionary to hold the models to be tested
models = {"Random Forest Classifier": rf,
          "Cat Boost Classifier": cat,
          "XG Boost Classifier": xgb}
```

```python
# Training and scoring the models
model_scores, predictions = fit_score(models, X_train, y_train, X_test, y_test)
```
